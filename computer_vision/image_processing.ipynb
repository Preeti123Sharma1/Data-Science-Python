filters

```python
filters =[
    ImageFilter.BLUR,
    ImageFilter.CONTOUR,
    ImageFilter.DETAIL,
    ImageFilter.EDGE_ENHANCE,
    ImageFilter.EDGE_ENHANCE_MORE,
    ImageFilter.EMBOSS,
    ImageFilter.FIND_EDGES,
    ImageFilter.SMOOTH,
    ImageFilter.SMOOTH_MORE,
    ImageFilter.SHARPEN
]
for i, f in enumerate(filters):
    im2.convert('RGB').filter(f).save(f'generated/filter_{i}.png')
    #im2.convert('RGB').filter(f).save(f'generated/filter_{i}.png') in normal laptop we write
```

```python
im2.convert('RGB').filter(ImageFilter.MaxFilter(7)).show()
im2.convert('RGB').filter(ImageFilter.MinFilter(7)).show()
im2.convert('RGB').filter(ImageFilter.ModeFilter(7)).show()
im2.convert('RGB').filter(ImageFilter.MedianFilter(7)).show()
im2.convert('RGB').filter(ImageFilter.GaussianBlur(7)).show()
```
